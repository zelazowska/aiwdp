###Przepisy
**Penne alla vodka**

- *3 tbsp. butter*
- *1 shallot, minced*
- *2 cloves garlic, minced*
- *1/2 c. tomato paste*
- *1/2 tsp. crushed red pepper flakes*
- *2 tbsp. vodka*
- *Kosher salt*
- *1 lb. tubed pasta, such as penne or rigatoni*
- *1/2 c. heavy cream*
- *1/2 c. freshly grated Parmesan, plus more for serving*
- *Basil, for serving*

1. In a large skillet over medium heat, melt butter. Add shallot and garlic and cook, stirring frequently, until softened, 4 to 5 minutes.
2. Add tomato paste and red pepper flakes and cook, stirring frequently, until paste has coated shallots and garlic and is beginning to darken, 5 minutes.
3. Add vodka to pot and stir to incorporate, scraping up any browned bits from the bottom of the pot. Turn off heat.
4. Bring a large pot of salted water to a boil and cook pasta until al dente. Reserve 2 cups of pasta water before draining.
5. Return sauce to medium heat and add 1/4 cup of pasta water and heavy cream, stirring to combine. Add half the Parmesan and stir until melted. Turn off heat and stir in cooked pasta. Fold in remaining Parmesan, adding more pasta water (about a tablespoon at a time) if the sauce is looking dry. Season with salt if needed. Serve topped with more Parmesan and torn basil leaves.

```python
# Zad 10

import numpy as np

operations = ["+", "-", "*", "/"]

while True:
  a = int(input("Pierwsza liczba: "))
  operation = input("Operacja: [+, -, *, /, log, sin, sqrt]: ")
  if operation in operations:
    b = int(input("Druga liczba: "))

    print(f"Twoje dzialanie: {a} {operation} {b} = ", end = "")
    if operation == "+":
      print(f"{a+b}")
    elif operation == "-":
      print(f"{a-b}")
    elif operation == "*":
      print(f"{a*b}")
    elif operation == "/":
      print(f"{a/b}")
    else:
      print("Niepoprawne dane")

  else:
    if operation == "log":
      print(f"Logarytm dziesietny liczby {a}: {np.log10(a)}")
    elif operation == "sin":
      print(f"Sinus liczby {a}: {np.sin(a)}")
    elif operation == "sqrt":
      print(f"Pierwiastek z liczby {a}: {np.sqrt(a)}")
    else:
      print("Niepoprawne dane")


  next_calculation = input("Czy chcesz wykonac kolejne dzialanie? (y/n): ")
  if next_calculation == "n":
    break
```

```python
# Zad 6

from math import sqrt

def is_prime(n):
  for i in range(2, int(math.sqrt(n)+1)):
    if n%i == 0:
      return "Nie jest pierwsza"
      break
  else:
    return "Jest pierwsza"

print(is_prime(137))
```

```python
# Zad 7

def is_integer(numbers):
  integers = []
  floats = []
  for number in numbers:
    if isinstance(number, int):
      integers.append(number)
    else:
      floats.append(number)
  if len(floats) > 0:
    return f"Liczby niecalkowite to: {floats}"
  else:
    return "Wszystkie liczby sa calkowite"

print(is_integer([1, 3, 5]))
print(is_integer([1.5, 2, 3]))
```

```python
# Zad 8

import math

def square_area(a):
  return f"Pole kwadratu: {a**2}"

def triangle_area(a, h):
  return f"Pole trójkąta: {0.5*a*h}"

def circle_area(r):
  return f"Pole koła: {round((math.pi*(r**2)), 2)}"


figure = input("Czego pole chcesz policzyc?\nkwadrat\ntrojkat\nkolo\n")
if figure.lower() == "kwadrat":
  a = int(input("Podaj dlugosc boku: "))
  print(square_area(a))

elif figure.lower() == "trojkat":
  a = int(input("Podaj dlugosc podstawy: "))
  h = int(input("Podaj wysokosc: "))
  print(triangle_area(a, h))

else:
  r = int(input("Podaj dlugosc promienia: "))
  print(circle_area(r))

```

```python
# Zad 3

import matplotlib.pyplot as plt
import pandas as pd

path = "https://drive.google.com/uc?export=download&id=1FEnUD6owTW9fH1JMBJWflfEPCGAe6p5o"
df = pd.read_csv(path)

plt.rc("font", size = 14)
plt.figure(figsize = (9, 7))
plt.title("Health points per Generation", fontsize=20)
plt.xlabel("HP", fontsize=14)

colors = ['yellow', 'orange', 'red', 'pink', 'purple', 'blue']

for generation in range(1, 7):
    data = df[df['Generation'] == generation]['HP']
    plt.hist(data, bins=10, edgecolor='black', alpha=0.7, color=colors[generation-1], label=f"Generation {generation}")

plt.legend()
plt.show()
```

```python
# Zad 4

import matplotlib.pyplot as plt
import pandas as pd

path = "https://drive.google.com/uc?export=download&id=1FEnUD6owTW9fH1JMBJWflfEPCGAe6p5o"
df = pd.read_csv(path)

ordinary = df[df["Legendary"] == False].groupby("Generation").size()
legendary = df[df["Legendary"] == True].groupby("Generation").size()

plt.rc("font", size = 14)
fig, ax = plt.subplots(figsize=(8, 8))
plt.title("Legendary and Ordinary Pokemons per generation", fontsize=16)
plt.xlabel("Generation")
plt.ylabel("Number of Pokemons")

plt.grid(True, linestyle='-', alpha=0.5)
plt.bar(ordinary.index, ordinary, width=1, label="Ordinary", color="steelblue", edgecolor="black")
plt.bar(legendary.index, legendary, width=1, label="Legendary", color="darkorange", edgecolor='black')
plt.legend(loc="center left")

plt.show()
```

```python
# Zad 5

import matplotlib.pyplot as plt
import pandas as pd

path = "https://drive.google.com/uc?export=download&id=1FEnUD6owTW9fH1JMBJWflfEPCGAe6p5o"
df = pd.read_csv(path)

plt.rc("font", size=14)
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(8, 10))
generation_labels = ['Generation 1', 'Generation 2', 'Generation 3', 'Combined generations']
colors = ['blue', 'green', 'red']

generation1 = df[df["Generation"] == 1]["HP"]
generation2 = df[df["Generation"] == 2]["HP"]
generation3 = df[df["Generation"] == 3]["HP"]

axes[0].hist(generation1, color='blue')
axes[0].set_title(generation_labels[0])

axes[1].hist(generation2, color='blue')
axes[1].set_title(generation_labels[1])

axes[2].hist(generation3, color='blue')
axes[2].set_title(generation_labels[2])

all_gen_hp = [generation1, generation2, generation3]
axes[3].hist(all_gen_hp, bins=10, color=['skyblue', 'orange', 'green'], label=['Gen 1', 'Gen 2', 'Gen 3'])
axes[3].set_xlabel('HP')
axes[3].set_ylabel('Count')
axes[3].set_title('Combined HP Distribution')
axes[3].legend()

plt.tight_layout()
plt.show()
```

```python
# Zad 6

import matplotlib.pyplot as plt
import pandas as pd

path = "https://drive.google.com/uc?export=download&id=1FEnUD6owTW9fH1JMBJWflfEPCGAe6p5o"
df = pd.read_csv(path)

plt.style.use("dark_background")
plt.figure(figsize=(10, 8))

defense = df["Defense"]
attack = df["Attack"]
total = df["Total"]

plt.scatter(defense, attack, c = total, s = 200, cmap = "plasma", alpha = 0.9)
plt.colorbar(label = "Total")

plt.xlabel("Defense")
plt.ylabel("Attack")
plt.title("Pokemon Strength")

plt.tight_layout()
plt.show()
```

```python
# Zad 7

import matplotlib.pyplot as plt
import pandas as pd

path = "https://drive.google.com/uc?export=download&id=1FEnUD6owTW9fH1JMBJWflfEPCGAe6p5o"
df = pd.read_csv(path)

generation_colors = {1: "purple", 2: "blue", 3: "brown", 4: "gray", 5: "cyan", 6: "magenta"}

plt.figure(figsize=(10, 8))

for gen, color in generation_colors.items():
    generation_data = df[df["Generation"] == gen]
    defense = generation_data["Defense"]
    attack = generation_data["Attack"]
    total = generation_data["Total"]

    plt.scatter(defense, attack, c=color, s=total, alpha=0.9, label=f'Generation {gen}')

plt.xlabel('Defense')
plt.ylabel('Attack')
plt.title("Strength")

plt.tight_layout()
```

```python
# Zad 8

import matplotlib.pyplot as plt
import pandas as pd

path = "https://drive.google.com/uc?export=download&id=1FEnUD6owTW9fH1JMBJWflfEPCGAe6p5o"
df = pd.read_csv(path)

legendary_color = "pink"
non_legendary_color = "yellow"

plt.figure(figsize=(10, 8))

non_legendary_data = df[df["Legendary"]==False]
non_legendary_defense = non_legendary_data["Defense"]
non_legendary_attack = non_legendary_data["Attack"]
plt.scatter(non_legendary_defense, non_legendary_attack, c=non_legendary_color, alpha=0.6, s=200, label="Non-Legendary")

legendary_data = df[df["Legendary"]==True]
legendary_defense = legendary_data["Defense"]
legendary_attack = legendary_data["Attack"]
plt.scatter(legendary_defense, legendary_attack, c=legendary_color, alpha=0.6, s=200, label="Legendary")

plt.xlabel("Defense")
plt.ylabel("Attack")
plt.title("Pokemon Strength")
plt.legend()
plt.xlim(-50, 250)
plt.ylim(-50, 200)

plt.tight_layout()
plt.show()
```

```python
# Zad 9

import matplotlib.pyplot as plt
import pandas as pd

path = "https://drive.google.com/uc?export=download&id=1FEnUD6owTW9fH1JMBJWflfEPCGAe6p5o"
df = pd.read_csv(path)
plt.figure(figsize = (10, 10))

data = df[df["Type 1"] == "Steel"]
total = data["Total"]
hp = data["HP"]
names = data["Name"]

plt.scatter(total, hp, color = "blue", s=20)
for i in range(len(total)):
    plt.annotate(names.iloc[i], (total.iloc[i], hp.iloc[i]), fontsize=8)

plt.xlabel("Total")
plt.ylabel("Health Points")
plt.title("Steel Type Pokemon")

plt.xlim(200, 800)
plt.ylim(30, 110)
plt.tight_layout()
plt.show()
```
